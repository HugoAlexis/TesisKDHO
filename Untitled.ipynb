```python
import matplotlib.pyplot as plt
import matplotlib.style as stl
import matplotlib.ticker as ticker
import matplotlib.colors
stl.use('seaborn')
import pandas as pd

from DampedKHO import *
```

```python
t_system = DampedKHO(5000, kappa=-3.5, kbt_ext=15, omega=1)
t_system.make_n_kicks(50)
```

```python
t_system.gamma = 0
t_system.kappa = 0

while t_system.omega > 1:
    t_system.omega -= 0.01
    t_system.make_n_kicks(1)
```

```python
# Get the energy evolution
e_bef, e_aft = t_system.get_energy_evolution()
e_mean = [(ei_bef + ei_aft)/2 for ei_bef, ei_aft in zip(e_bef, e_aft)]

# Join the Energy before and after the kicks
e_kicks = [*zip(e_bef, e_aft)]
energy = []
x = []
for index, e_kick in enumerate(e_kicks):
    energy.extend(e_kick)
    x.extend([index+1, index+1])
    
# Plot the energy evolution
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(range(1, len(e_bef) + 1), e_bef, color='blue', marker='o', markersize=6, linestyle='',
        label='$\langle E \\rangle_{\\rm antes}$')
ax.plot(range(1, len(e_bef) + 1), e_aft, color='red', marker='o', markersize=6, linestyle='',
        label='$\langle E \\rangle_{\\rm despues}$')
ax.plot(x, energy, linestyle='-',color='black')
ax.set(ylim=(0, 70), xlim=(0, 180))
ax.set_xlabel('Kick', fontsize=26, fontweight='bold')
ax.set_ylabel('$ \langle E \\rangle $', fontsize=26, fontweight='bold')
ax.tick_params(axis='both', which='major', labelsize=20)
ax.legend(loc='lower right', fontsize=20, ncol=1, facecolor=(0.9, 0.9, 0.9))

# Print the mean energy
print(np.mean(energy[-10:]))
plt.show()
```

```python
def plot_phase_space(ax, kappa, kbt, gamma, particles=400, kicks=200, printx=False, printp=False):
    t_system = DampedKHO(particles, kappa, kbt, gamma=gamma)
    t_system.make_n_kicks(kicks)

    ps_evolution, ps_evolution2 = t_system.get_phase_space_evolution()

    for ps in ps_evolution:
        ax.scatter(ps[:, 0], ps[:, 1], s=1, color='black')
    
    ax.set_title(f'$\kappa={kappa}$, $K_BT={kbt}$', fontsize=18,)
    ax.set(xlim=(-5, 5), ylim=(-5, 5), facecolor='white')
    ax.grid(False)
    ax.spines['bottom'].set(linewidth=3, color='black')
    ax.spines['right'].set(linewidth=3, color='black')
    ax.spines['left'].set(linewidth=3, color='black')
    ax.spines['top'].set(linewidth=3, color='black')
    if printx:
        ax.set_xlabel('x', fontsize=20, fontweight='bold', color='black')
    if printp:
        ax.set_ylabel('p', fontsize=20, fontweight='bold', color='black')
    ax.tick_params(axis='both', labelsize=18)
    ax.xaxis.set_major_locator(ticker.MaxNLocator(5))
    ax.yaxis.set_major_locator(ticker.MaxNLocator(5))
    
```

```python
fig, ax = plt.subplots(3, 3, figsize=(17, 14), sharex=True, sharey=True)
plot_phase_space(ax[0, 0], -0.5, 0, 0, printp=True)
plot_phase_space(ax[0, 1], -1.5, 0, 0)
plot_phase_space(ax[0, 2], -3.5, 0, 0)
plot_phase_space(ax[1, 0], -0.5, 0, 0.1, printp=True)
plot_phase_space(ax[1, 1], -1.5, 0, 0.1)
plot_phase_space(ax[1, 2], -3.5, 0, 0.1)
plot_phase_space(ax[2, 0], -0.5, 5, 0.1, printp=True, printx=True)
plot_phase_space(ax[2, 1], -1.5, 5, 0.1, printx=True)
plot_phase_space(ax[2, 2], -3.5, 5, 0.1, printx=True)
plt.show()
```
