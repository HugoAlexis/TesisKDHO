# **Oscilador armonico pateado**

## **Efecto de las patadas**

La dinamica del sistema esta descrito por el Hamiltoniano 
$$H = \frac{P^2}{2M} + \frac{m\omega^2 X^2}{2} + \frac{\kappa}{\sqrt{2}\eta^2} \cos{(\sqrt{2}\eta x)} \sum_n \delta(t-n\tau)$$

Tomamos las variables X y P como:
$$X = \sqrt{\frac{\hbar}{m\omega}}x, \verb|      | P=\sqrt{\hbar m \omega}p,$$

obtenemos,

$$H = \hbar\omega \left( \frac{p^2}{2} + \frac{x^2}{2} \right) + \frac{\kappa}{\sqrt{2}\eta^2} \cos\left(\sqrt{ \frac{2\hbar}{m\omega}}\eta x \right).$$

Utilizamos las ecuaciones de Hamilton para encontrar $\Delta p$ durante la n-esima patada.

$$\dot{x} = \frac{\partial H}{\partial p}, \verb|      | \dot{p} = -\frac{\partial H}{\partial x}.$$

$$\Rightarrow \verb|     | \dot{p} = -\hbar\omega x + \frac{\kappa}{\eta} \sqrt{\frac{\hbar}{m\omega}} \sin\left( \sqrt{\frac{2\hbar}{m\omega}}\eta x \right) \sum_n \delta(t-n\tau)$$

Resolvemos para el intervalo de $t = n\tau - \epsilon$ a $t = n\tau + \epsilon$ y obtenemos el limite $\epsilon \rightarrow 0$ 

$$\Delta p = \int_{n\tau - \epsilon}^{n\tau - \epsilon} \left[-\hbar\omega x + \frac{\kappa}{\eta} \sqrt{\frac{\hbar}{m\omega}} \sin\left( \sqrt{\frac{2\hbar}{m\omega}}\eta x \right) \sum_n \delta(t-n\tau) \right]dt$$

considerando que

$$\int_{n\tau - \epsilon}^{n\tau + \epsilon} \sum_n \delta(t - n\tau) dt= 1$$

obtenemos:

$$\Delta p = \frac{\kappa}{\eta}\sqrt{\frac{\hbar}{m\omega}} \sin \left(\sqrt{\frac{2\hbar}{m\omega}} \eta x \right)$$

## **Efecto del reservorio termico**

Entre las patadas, la dinamica del sistema esta descrito por la ecucion de Langevin:

$$\frac{d^2 X}{dt^2} + \frac{\gamma}{m}\frac{dX}{dt} + \omega^2 X = \frac{1}{m}\eta_{T}$$.

Haciendo el cambio de variables anterior, obtenemos:

$$\frac{d^2 x}{dt^2} + \frac{\gamma}{m}\frac{dx}{dt} + \omega^2 x = \frac{1}{m}\sqrt{\frac{m\omega}{\hbar}}\eta_{T}$$.

Podemos escribirlas como un sistema de ecuaciones diferenciales de primer orden:

$$
\begin{bmatrix}
x \\ p
\end{bmatrix}' = 
\begin{bmatrix}
0 & 1 \\
-\omega^2 & -\frac{\gamma}{m}
\end{bmatrix}
\begin{bmatrix}
x \\ p
\end{bmatrix} 
+
\begin{bmatrix}
0 \\ \eta_T\sqrt{\frac{\omega}{m\hbar}}
\end{bmatrix}.
$$

O bien, considerando $n$ particulas:
$$
\begin{bmatrix}
x_1 & p_1 \\
x_2 & p_2 \\
\vdots & \vdots \\
x_n & p_n
\end{bmatrix}' = 
\begin{bmatrix}
x_1 & p_1 \\
x_2 & p_2 \\
\vdots & \vdots \\
x_n & p_n
\end{bmatrix}
\begin{bmatrix}
0 & - \omega^2\\
1 & -\frac{\gamma}{m} 
\end{bmatrix} +
\begin{bmatrix}
0 &  \eta_T\sqrt{{\omega}/{(m\hbar})}  \\
0 &  \eta_T\sqrt{{\omega}/{(m\hbar})} \\
\vdots & \vdots \\
0 &  \eta_T\sqrt{{\omega}/{(m\hbar})}
\end{bmatrix}
$$

```python
import matplotlib.pyplot as plt
from DampedKHO import *
```

De acuerdo con el Hamiltoniano, la energir promedio del sistema es

$$\langle E \rangle = \frac{1}{N} \sum_{n=1}^N \omega\left[ \frac{p^2}{2} + \frac{x^2}{2} \right] \hbar.$$

Para la patada $n$, hay una energia $\langle E \rangle_{\rm antes}$ justo antes de la patada y otra justo despues de esta, $\langle E \rangle_{\rm despues}$.

```python
t_system = DampedKHO(10000, kappa=0, kbt_ext=15, omega=1)
t_system.make_n_kicks(50)
```

```python
# Get the energy evolution
e_bef, e_aft = t_system.get_energy_evolution()

# Join the Energy before and after the kicks
e_kicks = [*zip(e_bef, e_aft)]
energy = []
x = []
for index, e_kick in enumerate(e_kicks):
    energy.extend(e_kick)
    x.extend([index+1, index+1])
    
# Plot the energy evolution
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(range(1, len(e_bef) + 1), e_bef, color='blue', marker='o', markersize=6, linestyle='',
        label='$\langle E \\rangle_{\\rm antes}$')
ax.plot(range(1, len(e_bef) + 1), e_aft, color='red', marker='o', markersize=6, linestyle='',
        label='$\langle E \\rangle_{\\rm despues}$')
ax.plot(x, energy, linestyle='-',color='black')
ax.set(ylim=(0, 50), xlim=(0, 50))
ax.set_xlabel('Kick', fontsize=26, fontweight='bold')
ax.set_ylabel('$ \langle E \\rangle $', fontsize=26, fontweight='bold')
ax.tick_params(axis='both', which='major', labelsize=20)
ax.legend(loc='lower right', fontsize=20, ncol=1, facecolor=(0.9, 0.9, 0.9))

# Print the mean energy
print(np.mean(energy[-10:]))
plt.show()
```
